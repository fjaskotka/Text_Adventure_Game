```python
#Created by Fabian Jaskotka in 2020. 
#Pllease credit the author if willing to distribute

"""
Created on Thu Oct 13 2020 at 14:30

@author: Fabian Jaskotka 
"""


"""
DocString:

    A) Introduction:
        Normal flights without wind can be very romantic, but for the most
        part are rather boring. Flying is seen as the safest way to get from A to B. 
        Some believe that this is because it generally takes at least 2 unfortunate 
        happenings to get a plane down from the sky. 
        One of these happenings will mark the beginning of this game, a dead pilot...
        
        In 3 Levels, the players need to try to save the passengers and bring the
        plane down safely:
        
        Level 1: Players will be confronted with the captain and shall do him a 
                 deadly favor. Keep an item and present it to the captain.
                 
        Level 2: Players shall fly the plane based on commands of air traffic 
                 control with limited reaction time.
                 
        Level 3: Players need to find a helper, open a locked door and complete 
                 the landing checklist. 
                 
    B) Known Errors/Bugs:
        None. 

"""



#Importing necessary packages
import time
import random as rand

#Setting global objects
waiting_time = 1
passenger_no = 381

#Creating timestamp to calculate playing time
time_beginn = time.time()


#Listing of potential names for the Captain 
potential_names = ('George', 'Peter','Mike', 'Jack', 'Paul')

#Randomly choosing and assigning a name for the captain
cpt_name = rand.choice(potential_names)

#Checking the name that is chosen for the captain
#print(cpt_name)
 

##############################################################################
###########        ###########################################################
###        GREETING      ####
#############################
def welcome ():
    
    #Explanatory welcome text to give context to the game for the player
    print(f"""
   ___     ______      __   __                                     _ __    ___                  __ 
  / _ |   / _/ (_)__ _/ /  / /_  __ _____  __ __  _    _____  ___ ( ) /_  / _/__  _______ ____ / /_
 / __ |  / _/ / / _ `/ _ \/ __/ / // / _ \/ // / | |/|/ / _ \/ _ \|/ __/ / _/ _ \/ __/ _ `/ -_) __/
/_/ |_| /_//_/_/\_, /_//_/\__/  \_, /\___/\_,_/  |__,__/\___/_//_/ \__/ /_/ \___/_/  \_, /\__/\__/ 
               /___/           /___/                                                /___/          

    
    

{40 * '#'}
    WELCOME TO YOUR ADVENTURE OF TODAY!!
{40 * '#'}
    
    You are sitting in your comfortable economy class seat on a flight from JFK (New York)
    to SFO (San Francisco). A stewardess is approaching your seat and asks you to follow 
    her for a chat. You join her. In the galley, she tells you that one of the 
    pilots had a heart attack and that the other pilot wouldn't be able to land 
    the plane on his own - they have requested you to join them in the cockpit. 
    
    Are you ready for this challenge?""")
    
    #Waiting for the user to finish reading and confirm beginning of the game
    input(prompt="\n Pleass press <enter> to begin the game \n")
    
    #Calling function to begin the game
    level_1()
    
    
##############################################################################
###########        ###########################################################
###        LEVEL 1       ####
#############################
def level_1():
    
    #Defining global objects
    global name
    
    #Fitting the player with the chocolate bar
    chocolate_bar = 1
    
    #Welcoming in Level 1
    print(f"""



{40 * '#'})
    Level 1
{40 * '#'}""")
    
    #Asking for the chocolate bar 
    print(f"""
    You are carrying a chocolate bar in your hand.
    On your way to the cockpit a first-class passenger nicely asks you if he 
    could have it as there would be no more food on the plane. He says, he was starving.""")
    
    
    #Creating while loop to give player a chance to re-enter his choice in case it is not clear
    while True:
        
        #Asking the player to enter his choice
        chocolate_bar_quest = input(prompt="\nWould you like to give your chocolate bar to the other passenger?\n").lower()

        #Checking if the player wants to give his chocolate bar to the passenger, not accounting for undecicive people
        if (('yes' in chocolate_bar_quest or 'yeah' in chocolate_bar_quest or\
           'absolutely' in chocolate_bar_quest or 'sure' in chocolate_bar_quest or\
           'course' in chocolate_bar_quest) and not 'not' in chocolate_bar_quest) or\
           'why not' in chocolate_bar_quest:
            chocolate_bar -= 1
            time.sleep(waiting_time)
            print("""
    Great, you made someone very happy! ðŸ˜ƒ""")
            break

        #Checking if the player intentionally does not want to pass on his chocolate bar
        elif 'no' in chocolate_bar_quest:
            print("""
    Your choice! You might even need it, who knows...
            """)
            break

        #Asking the user to repeat their entry more clearly
        else:
            print("""
    I'm having a hard time hearing you.
    Please make your choice if you would like to give the chocolate bar more clearly.""")

    
    #Going into the cockpit
    time.sleep(waiting_time * 2)
    print(f"""\n {20 * '#'}
    
    +---------------------------------+
    | You have made it to the cockpit.|
    +---------------------------------+
    """)
    time.sleep(waiting_time)
    
    
    #Introduction by the captain
    print(f"""
    ____________________________________________________________________________
    |Captain: Thanks for coming over! I heard you are willing to help me try to  
    |         save all {passenger_no} people onboard this aircraft.                           
    |         I am {cpt_name}, what's your name?                                     
    |___________________________________________________________________________
    """)                            
    time.sleep(waiting_time * 2)
    
    
    #Creating loop to make people re-enter their name in case they introduce themselves in sentences
    while True:
    
        #Asking for the name of the player
        name = input(prompt="\nPlease tell the captain your name!\n").lower().title()

        #Checking, if player wrote a text with commas or a sentence with more than 4 words
        if "," in name or len(name.split()) > 4:
            print("""
        This is not a real name. Please do not write sentences, but just your name.
        """)
            continue
        
        #Checking if the player introduces herself / himself with 'I am'
        elif 'I Am' in name:
            words = name.split()
            name = words[2]
            break

        #Checking if the player introduces herself / himself with "I'm"
        elif "I'M" in name:
            words = name.split()
            name = words[1]
            break

        #Checking if the player introduces herself / himself with 'My name is'
        elif 'Name Is' in name:
            words = name.split()
            name = words[3]
            break
        
        #Checking if the name is really just one word
        elif len(name.split()) == 1:
            break
            
        #Catching errors
        else:
            print("Something went wrong in the identification of your name. Please try again with your first name.")
            continue

        

    time.sleep(waiting_time)
    
    #Printing quest by the captain for the chocolate bar
    print(f"""
    ______________________________________________________________________________________
    |Captain: Well, alright then {name}, before we get into flying, I have to tell you that 
    |         I am starving...
    |_____________________________________________________________________________________
    """)
    time.sleep(waiting_time * 3)
    print("""
    ____________________________________________
    |Captain: Do you have a little snack for me?
    |___________________________________________
    """)
    time.sleep(waiting_time * 1.5)
    
    
    #Creating a loop to alow the player to re-enter his choice in case it is not easy to identify the intention
    while True:
        
        #checking if the player holds the chocolate bar
        if chocolate_bar < 1:
            snack = input(prompt="""
How do you want to answer to the captain?
    1) Tell the captain, you do not have one
    2) Tell the captain, you might be able to get him one
    3) Ask the question, if it was important
""")  
        
        #answer possibilities in case the player does not have the chocolate bar
        else:
            snack = input(prompt="""
How do you want to answer to the captain?
    1) Give the chocolate bar to the captain
    2) Since you are hungry yourself, tell him you don't have anything
    3) Don't give your bar to the captain and instead eat it yourself
""")
        #General waiting until the captain responds
        time.sleep(waiting_time)
        
        #Checking if player does not have the chocolate bar and tels the captain that he doesn't have one        
        if chocolate_bar < 1 and ('1' in snack or 'not' in snack.lower() or 'not have' in snack.lower() or "don't have" in snack.lower()):
            
            #Sending the player back to the beginning of this chapter to not give away the chocolate bar
            print("""
    The captain nicely asks you to see if you can find something!
    """)
            time.sleep(waiting_time)
            input(prompt="Press <enter> to go back to the beginning of this chapter")
            level_1()
            break
        
        #Checking if player does not have the chocolate bar but offers captain to get one
        elif chocolate_bar < 1 and ('2' in snack or 'get' in snack.lower() or 'able' in snack.lower()):
            
            #Printing the captain's appreciation
            print("""
    The captain loves your effort and says that it is fine. It was just an engagement check.
    """)
            time.sleep(waiting_time * 2)
            
            #Printing the death of the captain and sending to level 2
            print(f"""
    The captain's head falls down, his eyes remain open. He collapsed...
    After a minute you are realizing that you are the one in charge of {passenger_no} people now.
            """)
            level_2()
            break
            
        #Checking of player does not have the chocolate bar and questions importance of quest
        elif chocolate_bar < 1 and ('3' in snack or 'question' in snack.lower() or 'important' in snack.lower()) :
            
            #Printing captain's anger and sending back to the seat --> loosing the game
            print("""
    The captain is not amused and kicks you out of the cockpit for your rude behavior of questioning his authority.
    """)
            time.sleep(waiting_time)
            print("""
    You have to take a seat in your economy class seat again and cannot do anything anymore to safe the plane.
    """)
            loose()
            break
            
        #Checking if player holds the chocolate bar and wants to give it to the captain
        elif chocolate_bar >= 1 and (('1' in snack or 'give' in snack.lower() or ('to' in snack.lower() and 'captain' in snack.lower()) or 'chocolate' in snack.lower() or 'sure' in snack.lower() or 'absolutely' in snack.lower()) and not 'not' in snack.lower()):
            
            #Printing the captain's appreciation and his following death --> sending to level 2
            print(f"""
    _____________________________________________________        
    |Captain: Thank you so much, {name}! It is delicious
    |____________________________________________________
    """)
            time.sleep(waiting_time * 2)
            print(f"""
    The captain's head falls down, his eyes remain open. 
    After a minute you are realizing that the chocolate bar was poisoned 
    and that you are the one in charge of {passenger_no} people now.
            """)
            level_2()
            break
            
        #Checking if plaser holds the chocolate bar but tells the captain he doesn't
        elif chocolate_bar >= 1 and ('2' in snack or 'hungry' in snack.lower() or ('not' in snack.lower() and 'have' in snack.lower()) or 'anything' in snack.lower()):
            
            #Making the captain kick the player out of the cockpit --> loosing the game
            print("""
    The captain finds out only a short time later and kicks you out of the cockpit. 
    He doesn't work with liars, he says.
    """)
            time.sleep(waiting_time)
            print("""
    You have to take a seat in your economy class seat again and watch the captain loose control.
    """)
            loose()
            break
            
        #Checking if the player holds the chocolate bar but wants to eat it himself/herself
        elif chocolate_bar >= 1 and ('3' in snack or ('not' in snack.lower() and 'give' in snack.lower()) or 'eat' in snack.lower()):
            
            #Simulating the death of the player due to the poisoning --> loosing the game
            print("\n\nMhh, that felt good")
            time.sleep(waiting_time)
            print("\nWait a second...")
            time.sleep(waiting_time)
            print("\nWhat is happening?")
            time.sleep(waiting_time)
            print("\nYou feel your eyes closing and head falling... ")
            time.sleep(waiting_time)
            print("\nThe chocolate bar was poisoned!! ")
            time.sleep(waiting_time)
            print("\nNow the captain has no chance to land the plane safely...")
            time.sleep(waiting_time)
            loose()
            break
            
        #Checking if generally the player answers intuitively with the intention to not give anything to the captain
        elif 'no' in snack.lower():
            
            #Printing captain's anger and sending back to the seat --> loosing the game
            print("""The captain is not amused and kicks you out of the cockpit for your rude behavior of questioning his authority.""")
            time.sleep(waiting_time)
            print("""you have to take a seat in your economy class seat again and watch the plane crash.""")
            loose() 
            break
            
        #Checking for unclear answers that are not accounted for above
        else:
            print("Sorry this answer was not easy to identify. Please formulate your answer more clearly among the examples.")
            time.sleep(waiting_time)                          
        
        
        
        
##############################################################################
###########        ###########################################################
###        LEVEL 2       ####
#############################
def level_2():
    
    #Letting the player time to realy before the next level starts
    time.sleep(waiting_time * 4)
    
    #Welcoming in Level 2
    print(f"""



{40 * '#'})
    Level 2
{40 * '#'}""")
    time.sleep(waiting_time)
    print("""
    Congrats, you are in charge of the plane's controls now. """)
    time.sleep(waiting_time)
    
    
    #Explaining how the next challenge works and how to fly the plane
    print("""
    
    Please navigate through the obstacles carefully! 
    The air traffic control will help you to not crash into something by giving 
    you hints how to behave.
    
    Your airplane controls are as follows:
    a = turning left
    d = turning right
    s = ascending
    w = decending
    
    IMPORTANT:  You only have 2 seconds to answer! 
                Always hit the <enter> button after your choice is made!
    
    """)
    
    #Giving the player time to read the instructions before confronting her/him with decisions to make
    #time.sleep(waiting_time * 5)
    
    #Creating a while loop to give the player a chance to re-enter his choice in case it is not clear.
    while True:

        #Asking the player to play the navigation game
        start_navigation = input(prompt="\nAre you ready to navigate this 280 tons metal tube through the skies?\n").lower()

        #Checking if the player is ready to play the game
        if ('yes' in start_navigation or 'yeah' in start_navigation or 'course' in start_navigation or 'sure' in start_navigation or 'absolutely' in start_navigation or 'definitely' in start_navigation) and not 'not' in start_navigation:
            fly()
            break

        #Checking if the player does not feel ready to play the game
        elif 'no' in start_navigation:
            print(f"""
            
    No worries, they will probably find someone else to do this for the {passenger_no} passengers on board.
    """)
            time.sleep(waiting_time)
            print("""
    Oh maybe not...
    """)
            time.sleep(waiting_time)
            loose()
            break
            
        #Checking for unclear answers that are not accounted for above
        else:
            print("""
    I'm having a hard tim hearing you.
    Please make your choice more clearly.""")


        
        
##############################################################################
### Flying the plane ##
#######################
def fly():

    #Waiting to let the user relax and get ready
    time.sleep(waiting_time * 2)
    
    #Initiating suspense at the beginning of the challenge
    print("""
    Great, let's get ready to fly
    """)
    
    #Setting length of countdown
    countdown = 5
    
    #Launching and printing countdown
    while countdown > 0:
        time.sleep(waiting_time)
        print(f"""
    {countdown}""")
        countdown -= 1
    
    
    #Defining scenarios for the air traffic control to confront the player with
    scenario = ["Turn left!", "Turn right!", "Decent!", "Pull up!"]
    
    #Setting the variable for how many rounds have been played
    round_no = 1
    
    #Setting the variable to count wrong reactions
    wrong_reaction = 0
    
    
    #Creating loop to go through the rounds and let the player react to the situations
    while round_no <= 10:
        
        #Creating a little relaxing time between the rounds
        time.sleep(waiting_time * 2)
        
        #Make the computer pick a random scenario to play
        random_choice = rand.randint(0,3)
        
        #Show the scenario to the player
        print(f"""\n {scenario[random_choice]}""")
        
        #Recording timestamp when the scenario is printed
        time_stamp_1 = time.time()
        
        #Making the user navigate the plane
        player_reaction = input(prompt=">").lower()
        
        #Recording timestamp when player inputted his reaction choice
        time_stamp_2 = time.time()
        
        #Calculating response time
        response_time = time_stamp_2 - time_stamp_1
        
        #Checking if the response time was not good enough and too many errors made
        if response_time > 2 and wrong_reaction > 2:
            print("""Oh no, that is getting close""")
            time.sleep(waiting_time)
            print("""We can't do anything anymore!!!""")
            time.sleep(waiting_time)
        
            #Calling loose function
            loose()
        
        #Checking if response time was not good enough
        elif response_time > 2:
            print("""Oh, that was not fast enough! Please do it faster next time!""")
            
            #Adding a wrong reaction
            wrong_reaction += 1
            
            #Going a round further
            round_no += 1
             
        
        #Checking if the answer of the player is correct
        elif random_choice == 0 and player_reaction == 'a' or random_choice == 1 and player_reaction == 'd' or random_choice == 2 and player_reaction == 'w' or random_choice == 3 and player_reaction == 's':
            
            #Checking if it is the last round
            if round_no == 10:
                print("""Great job!""")
                break
            
            #Checking if it was not the last ound
            elif round_no < 10:
                
                motivation_options = ['Great job! Keep going', 'Terrific, kepp it up!', 'Nice work! Stay focused']
                motivation = rand.choice(motivation_options)
                print(f"{motivation}")
                round_no += 1
             
            #Accounting for errors
            else:
                print("""Something went wrong in the identification of your current round number.""")
                
                #Ask player to fly again
                input(prompt="Press <enter> to fly again")
                fly()
        
        #Checking if the player hasn't made any mistakes yet
        elif wrong_reaction < 1: 
            print("""Wow, that was close, but we got lucky, nothing happened!
            Focus harder!
            """)
            wrong_reaction += 1
            round_no += 1
        
        #Checking if the player has made made her/his first mistake
        elif wrong_reaction <2:
            print("""Come on, you can do better than this!! Keep it together.""")
            wrong_reaction += 1
            round_no += 1
            
        #Letting players with too many wrong reactions fail    
        elif wrong_reaction >=2:
            print("""Oh no, that is getting close""")
            time.sleep(waiting_time)
            print("""We can't do anything anymore!!!""")
            time.sleep(waiting_time)
        
            #Calling loose function
            loose()
        
        #Checking for errors
        else:
            print("""Something went wrong in the identificaion with you reactions, please start the flying challenge again""")
            
             #Ask player to fly again
            input(prompt="Press <enter> to fly again")
            fly()
    
    #Waiting before success message
    time.sleep(waiting_time * 2)
    
    #Defining success message for great pilots with no mistakes before moving to level 3    
    if wrong_reaction == 0:
        print("""
        
                                   _             
         /\                       (_)            
        /  \   _ __ ___   __ _ _____ _ __   __ _ 
       / /\ \ | '_ ` _ \ / _` |_  / | '_ \ / _` |
      / ____ \| | | | | | (_| |/ /| | | | | (_| |
     /_/    \_\_| |_| |_|\__,_/___|_|_| |_|\__, |
                                            __/ |
                                           |___/ 

    Wow, top Job!! You are an excellent pilot!
    """)
        level_3()
     
    #Defining success message for good pilots with only 1 mistake before moving to level 3
    elif wrong_reaction <= 1:
        print("""
        
    Cool, you have made it! 
    It was close only 1 single time!
    """)
        level_3()
    
    #Defining success message for pilots tht made it before moving to level 3
    elif wrong_reaction <= 3:
        print("""
        
    Cool, you have made it! 
    It was close just a few times but you successfully managed to 
    navigate the plane around the obstacles ðŸ§‘â€âœˆï¸
    """)
        level_3()
    
    #Checking for errors that are not accounted for 
    else:
        print("""Oops, something went wrong!! Please fly again.""")
         
        #Ask player to fly again
        input(prompt="Press <enter> to fly again")
        fly()


        
        
##############################################################################
###########        ###########################################################
###        LEVEL 3       ####
#############################        
def level_3():
    
    #Letting the player relax before getting into the next challenges
    time.sleep(waiting_time * 4)
    
    #Welcoming in Level 3
    print(f"""



{40 * '#'})
    Level 3
{40 * '#'}""")
    time.sleep(waiting_time * 2)
    
    print("""
    Congratulations! You have made it to the final stage to save the passengers 
    onboard this plane.
    """)
    time.sleep(waiting_time * 2)
    
    #Explaining to the player that he will need a second person in the cockpit
    print("""
    To be able to land this plane, you will need a second person in the cockpit. 
    """)
    time.sleep(waiting_time)
    
    #Creating a loop to make sure either the first-class man or the flight attendant is selected
    while True:
        
        #Asking player for the desired company for the final stage 
        assistant = input(prompt="Who shall accompany you?\n").lower()

        #Checking if player means the flight attendant from the beginning of this adventure
        if 'steward' in assistant or 'attendant' in assistant or ('person' in assistant and 'beginning' in assistant):
            assistant = 'stewardess'
            break
        
        #Checking if player means the first-class person that wanted the chocolate bar
        elif 'first' in assistant or 'chocolate' in assistant:
            assistant = 'first-class man'
            break
        
        #Checking if the the player wants the computer to choose an assistant
        elif 'care' in assistant or ('not' in assistant and 'care' in assistant):
            print("""
    Alright, then a person will be chosen for you.""")
            time.sleep(waiting_time * 2)
            
            #Defining the list of possible helpers for the final stage
            assistant_options = ['flight attendant', 'first-class man']
            
            #Count for the suspense
            suspense = 3
            
            #Loop to print suspense statement and create...suspense
            while suspense > 0:
                suspense -= 1
                print("""
    choosing...""")
                time.sleep(waiting_time)
            
            #Choosing a random person as a helper from the list assistanassistant_options
            assistant = rand.choice(assistant_options)
            
            #Printing the choice to the player
            print(f"""
            
    The {assistant} from the beginning of this game has been chosen as your 
    helper to support you!
    """) 
            time.sleep(waiting_time)
            break
        
        #Checking if the player chooses the captain to help him
        elif 'captain' in assistant or f"{cpt_name}" in assistant:
            print(f""" 
    Unfortunately, you killed {cpt_name}. Therefore, he won't be able to help you anymore
    """)
        
        #Handling people that are desired, but not on this flight
        else:
            print(f"""
    Unfortunately, {assistant} is not on this flight. Please choose another helper!
    Hint: You only know 2 people on this plane that are still alive.
    """)

    #Leaving the cockpit to get the assistant
    time.sleep(waiting_time * 2)
    print(f"""
    +-------------------------------------------------------------------+
    |You are on your way to the main cabin to to get the {assistant}.   |
    +-------------------------------------------------------------------+
    """)
    
    #Prompting player with closed cockpit door
    time.sleep(waiting_time * 2)
    print(f"""
    On your way back to the cockpit you find out that the door is locked...
    """)
    
    #Defining limit of password tries
    password_try = 3
    
    #Setting password to number of passengers on the plane
    password = cpt_name
    
    #Creating a loop for password request of cockpit door
    while password_try > 0:
        time.sleep(waiting_time * 2)
        
        #Letting user try the password
        password_attempt = input(prompt="Please enter the password for the cockpit door.\n").lower().title()
        
        #Checking if password is correct
        if password_attempt == password:
            print("""
    Great, the door is open! You are now back in the cockpit""")
            break
        
        #Checking if password was wrong, then decreasing the tries left
        elif password_attempt != password:
            password_try -= 1
            print(f"""
    Sorry, that was not the right password
    You have {password_try} more chances to enter the right password.
    Hint: Remember the name of the captain from this plane""")
        
        #Catching errors in the password validation process
        else:
            print("Oops, something went wrong when trying to enter the password.")
            
            #Letting user restart this Level
            input("Press <enter> to go back to the beginning of level 3")
            level_3()
            break
        
    #Checking if no more password attempts are left and letting those players fail that did not manage to find out the right password
    if password_try == 0:
        print("""
    Oh no, the door is now electronically sealing itself. 
    Nobody will be able to enter anymore!
    Nobody, will be able to land this plane anymore!!""")
        loose()
    
     
    
    #Prompting the sucessors of the password challenge with the passenger announcement
    time.sleep(waiting_time * 2)
    print(f"""
    {assistant.title()}: Alright, look what I found:
    
    ______________________________________
    |                                    |
    |         LANDING CHECKLIST          |
    |             ---------              |
    |                                    |
    |    1. Make sure 2 people are   âœ… |
    |       in the cockpit.              |
    |                                    |
    |    2. Do a passenger               |
    |       announcement.                |
    |                                    |
    |____________________________________|
    
    """)
    time.sleep(waiting_time * 4)
    
    #Defining variable with chances to do propper announcement
    chances = 3
    
    #Creating loop to repeat announcement in case it does not match the criteria
    while chances > 0:
        
        #Asking for passenger announcement by player
        passenger_announcement = input(prompt="Please announce to the passengers that you will be landing shortly.\n")
        time.sleep(waiting_time * 2)
        
        passenger_announcement_wip = passenger_announcement.lower()
        
        #Checking if the announcement is a not sufficient one
        if 'ladies' not in passenger_announcement_wip and 'gentlemen' not in passenger_announcement_wip and 'land' not in passenger_announcement_wip and 'passenger' not in passenger_announcement_wip:
             
            #Checking if the player has more than one chance left  
            if chances > 1: 
                chances -= 1
                print(f"""
    This does not sound like a propper passenger announcement. Please try again.
    Hint: Make it sound like a real announcement from your past flight!

    You have {chances} chances left.
            """)
                
            #Checking if the player only has one chance left and giving one extra hint
            elif chances == 1:
                chances -= 1
                print(f"""
    This does not sound like a propper passenger announcement. Please try again.
    Hint: Make it sound like a real announcement from your past flight!
    
    EXTRA HINT: It should include the word 'passenger' and 'land'

    You have {chances} chances left.""")
                
            else:
                print("Something went wrong in the calculation of your remaining chances")
                
                #Letting user restart this Level
                input("Press <enter> to go back to the beginning of level 3")
                level_3()
                
      
        #Checking if the answer is a propper one as imagined by the designer
        elif 'ladies' in passenger_announcement_wip or 'gentlemen' in passenger_announcement_wip or 'land' in passenger_announcement_wip or 'passenger' in passenger_announcement_wip:
            break
        
        #Catching potential errors and situations that are not accounted for above
        else:
            print("""Something went wrong with you passenger announcement.""")
       
    
    #Defining text action when a player has no more chances left
    if chances == 0:
        print("""
    You have not managed to inform your passengers about your landing...
    """)
        time.sleep(waiting_time)
        print("""
    Nobody is on their seat when approaching the runway. 
    In fact, somebody even knocks on the cockpit door and scares you in the final moment.
    """)
        time.sleep(waiting_time)
        print("""
    You loose focus and ultimately also control of the plane....
    """)
        loose()
              
    
    #Playing passenger announcement of player
    print(f"""
    ____________________________________________________________________________
    |In the backgound: "{passenger_announcement}" 
    |___________________________________________________________________________
    """)
    time.sleep(waiting_time * 2)
    
    
    #Showing assistant's appreciation for the announcement
    print(f"""
    ___________________________________________________________
    |{assistant.title()}: Beautiful! Now everybody is informed!
    |__________________________________________________________
    """)
    time.sleep(waiting_time * 2)
    
    
    #Printing the updated checklist
    print("""
    
    ______________________________________
    |                                    |
    |         LANDING CHECKLIST          |
    |             ---------              |
    |                                    |
    |    1. Make sure 2 people are   âœ… |
    |       in the cockpit.              |
    |                                    |
    |    2. Do a passenger               |
    |       announcement.            âœ… |
    |                                    |
    |____________________________________|
    """)
        
    
    #Introducing the player to the final challenge, to bring the plane down safely
    time.sleep(waiting_time * 2)
    print(f"""
    ____________________________________________________________________________________
    |{assistant.title()}: Alright, {name}, now just bring this big bird onto the runway!
    |___________________________________________________________________________________
    """)
    time.sleep(waiting_time)
    
    #Setting plane's current altitude
    altitude = 500
    
    #simulating announcements of plane shortly before touchdown
    while altitude > 0:
        print(f"""
        +------------+
        |Plane: {altitude}  |
        +------------+
        """)
        altitude -= 100
        time.sleep(waiting_time * 1.5)
    
    #Asking the player to touch the plane down
    input(prompt="Press <enter> to touchdown ")
    
    #Calling the winning function
    win()
    
    

##############################################################################
###########        ###########################################################
###        WINNING       ####
############################# 
def win():
    
    #Creating animation of landing plane
    time.sleep(waiting_time * 1.5)
    print("""
    
                                  -------/
                                /       /
                               /       /
                             __       /
            ________________--       /__
       ___-                           __O
     __________<=======/\_________---
       |        0===>   | |
       O                O O""")
    
    #Set runway length
    runway = 75
    
    #Set start speed of the runway passing under the plane
    speed = 0.01
    
    #Making runway pass under the plane, with decreasing speed
    while runway > 0:
        runway -= 1
        print("%", end="")
        time.sleep(waiting_time * speed)
        speed *= 1.04
        
        
    #Congrats to the winning player
    print("""
    
oooooo   oooo                            oooooo   oooooo     oooo  o8o                   .o. 
 `888.   .8'                              `888.    `888.     .8'   `"'                   888 
  `888. .8'    .ooooo.  oooo  oooo         `888.   .8888.   .8'   oooo  ooo. .oo.        888 
   `888.8'    d88' `88b `888  `888          `888  .8'`888. .8'    `888  `888P"Y88b       Y8P 
    `888'     888   888  888   888           `888.8'  `888.8'      888   888   888       `8' 
     888      888   888  888   888            `888'    `888'       888   888   888       .o. 
    o888o     `Y8bod8P'  `V88V"V8P'            `8'      `8'       o888o o888o o888o      Y8P 
                                                                                                                         

    """)
    
    #Creating timestamp for finish of the game
    time_end = time.time()
    
    #Calculating and printing playing time
    playing_time = round(time_end - time_beginn)
    print(f"""
    It took you {round(playing_time // 60)} mins and {playing_time % 60} seconds to save everybody!
    """)
    
    #Asking if the player wants to play another round
    again = input(prompt="Do you want to play again? Yes or no?").lower()
    
    if ('yes' in again or 'absolutely' in again or 'sure' in again or 'definitely' in again) and not 'not' in again:
        welcome()
        
    else:
        exit(0)
        
        
        
##############################################################################
###########        ###########################################################
###        LOOSING       ####
#############################        

def loose():
    
    #Letting the looser relax before the final message 
    time.sleep(waiting_time * 1.5)
    
    #Letting the player know that her/his plane has crashed
    print("\n\n\nOH NOOOO, the plane crashed!!!!")
    
    #Printing the official loose statement
    print("""
         )             (                         
  ( /(             )\ )                      
  )\())       (   (()/(                  (   
 ((_)\  (    ))\   /(_))  (    (   (    ))\  
__ ((_) )\  /((_) (_))    )\   )\  )\  /((_) 
\ \ / /((_)(_))(  | |    ((_) ((_)((_)(_))   
 \ V // _ \| || | | |__ / _ \/ _ \(_-</ -_)  
  |_| \___/ \_,_| |____|\___/\___//__/\___|  
                                             

    """)
    print("""\nGAME OVER""")
    
    #Creating timestamp for finish of the game
    time_end = time.time()
    
    #Calculating and printing playing time
    playing_time = round(time_end - time_beginn)
    print(f"""
    You have played {round(playing_time // 60)} mins and {playing_time % 60} seconds.
    """)
    
    #Asking if the player wants to play another round
    again = input(prompt="Do you want to play again? Yes or no?").lower()
    
    if ('yes' in again or 'absolutely' in again or 'sure' in again or 'definitely' in again) and not 'not' in again:
        welcome()
        
    else:
        exit(0)
        

#Starting the game
welcome()
```
